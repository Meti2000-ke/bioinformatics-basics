```python
# seq record
# A seq record is an object that stores a biological sequence along with its details
# The details include things like: ID, name, and description
 
from Bio.SeqRecord import SeqRecord

dna = Seq("ATGCGT")

record = SeqRecord(
    dna,
    id = "Gene1",
    name = "Example Gene",
    description = "Sample DNA Seq for practice"
)

print(record.id)
print(record.name)
print(record.description)
print(record.seq)
```

```python
# FASTA files and SEQIO in Biopython

from Bio import SeqIO

record = SeqIO.read("COX1.fasta.fasta", "fasta")

print(record.id)
print(record.description)
print(len(record))
print(record.seq)
```

```python
#  If we want to analyze more than one sequence at once we use #parse function rather than read function

from Bio import SeqIO

for record in SeqIO.parse("COX1.fasta.fasta", "fasta"):
    print(record.id)
    print(record.description)
    print(len(record))
    print("-"*30)
```

```python
# the ("-"*30) code is we are actually telling it to print the - line 30 time if we insert "a" instead of "-" it will print the line aaaaaa 30 times to break between the sequences
```

```python
# how to deal with GenBank file format

from Bio import SeqIO

record = SeqIO.read("COX1.GenBank", "genbank")

print(record.id)
print(record.description)
print(record.annotations)
print(len(record.features))
```

```python
# if we want to see what is inside the future of that genbank file

for feature in record.features:
    print(feature.type)
```

```python
# if we also want to see the details of the future ...

for feature in record.features:
        print(feature.type , feature.location)

```

```python
# BLAST on BioPython 

from Bio.Blast import NCBIWWW

result_handle = NCBIWWW.qblast(
    program = "blastp",
    database = "nr",
    sequence = "MALWMRLLPLLALLALWGPDPAAAFVNQHLCGSHLVEALYLVCGERGFFYTPKTRREAEDLQVGQVELGGGPGAGSLQPLALEGSLQKRGIVEQCCTSICSLYQLENYCN"
)

with open ("blast_result.xml", "w") as b:                      # to open a file
    b.write(result_handle.read())

print("BLAST performed successfully!")
```

```python
# "nr" means that when u give us the value make it non redundant
# we use the function "with" to open a new file or document
# "blast_result.xml" is the name we assigned to the new document
# "w" means that open in "write" format or write mode rather than read only mode
# "b" is the nickname we are giving to the name we assigned.
# "b.write(result_handle.read())" this function means that I want everything written under result_handle to be in the "b" file and to file to file to be read
# its important that we save a blast file in .xml format
```

```python
# what if our sequence is long that copy and pasting it is crowded?

from Bio import SeqIO

record = SeqIO.read("Insulin human uniprotkb.fasta", "fasta")

result_handle = NCBIWWW.qblast(
    program = "blastp",
    database = "nr",
    sequence = record.seq
)
with open ("blast_result.xml", "w") as b:                      # to open a file
    b.write(result_handle.read())

print("BLAST performed successfully!")

# Then done!
```

```python
# how to read blast or (.xml) files

from Bio.Blast import NCBIXML

with open("blast_result.xml") as b:
    blast_record = NCBIXML.read(b)
```

```python
# to find out how many sequences are found in that blast file?

print(len(blast_record.alignments))
```

```python
# to access or print the first sequence we use indexing method

first_alignment = blast_record.alignments[0]
print(first_alignment.title)
print(first_alignment.length)
```
